```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]

img = cv2.imread('images/image.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(421), plt.imshow(img)

row, col = len(img)-1, len(img[0])-1
white = img[row, col, :]
coeffs = 255.0/white

patched = np.zeros_like(img, dtype=np.float32)
for channel in range(3):
    patched[..., channel] = img[..., channel] * coeffs[channel]

# White patching does not guarantee that the dynamic range is preserved, images must be clipped.
patched = patched/255
patched[patched > 1] = 1

plt.subplot(423), plt.imshow(img)
plt.subplot(424), plt.imshow(patched)

#Gray world
normalized_img=img/255
mean_r=np.average(normalized_img[:,:,0])
mean_g=np.average(normalized_img[:,:,1])
mean_b=np.average(normalized_img[:,:,2])
gray=(mean_r+mean_g+mean_b)/3
kr=gray/mean_r
kg=gray/mean_g
kb=gray/mean_b
print(gray,kr,kg,kb)
patched[:,:,0]=normalized_img[:,:,0]*kr
patched[:,:,1]=normalized_img[:,:,1]*kg
patched[:,:,2]=normalized_img[:,:,2]*kb
patched[patched > 1] = 1

plt.subplot(425), plt.imshow(img)
plt.subplot(426), plt.imshow(patched)

#max scale
max_r=max([ max(r) for r in patched[:,:,0]])
max_g=max([ max(r) for r in patched[:,:,1]])
max_b=max([ max(r) for r in patched[:,:,2]])

print(max_r,max_g,max_b)

maxscaled = np.zeros_like(img, dtype=np.float32)
maxscaled[:,:,0]=patched[:,:,0]/max_r
maxscaled[:,:,1]=patched[:,:,1]/max_g
maxscaled[:,:,2]=patched[:,:,2]/max_b
maxscaled[maxscaled > 1] = 1

plt.subplot(427), plt.imshow(patched)
plt.subplot(428), plt.imshow(maxscaled)
```
